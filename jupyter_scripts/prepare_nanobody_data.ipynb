```python
import pandas as pd
```

```python
raw_binding_data = pd.read_csv(r"C:\Users\nilsh\OneDrive\Desktop\results_thesis\data\Antibody bank.xlsx - VHH overview.csv", header = 1)
columns_to_keep = ["Full sequence", "Selected on", "Ag1", "Ag1_Raw signals", "Name VHH", "Selection", "CDR3", "Framework 3", "CDR2", "Framework 2", "CDR1", "Framework 1"]
tidied_data = raw_binding_data[columns_to_keep]
tidied_data.head(20)
```

Normalize raw signals based on group in Selection


```python
# Apply normalization to sum to 1 within each group
# Normalize the column so that the max value becomes 1
tidied_data['Normalized_Ag_Raw_signals'] = tidied_data['Ag1_Raw signals'] / tidied_data['Ag1_Raw signals'].max()
# Normalize 'Ag_Raw_signals' such that the maximum value in each group is 1
tidied_data['Normalized_Ag_Raw_signals'] = tidied_data.groupby('Selection')['Ag1_Raw signals'].transform(lambda x: x / x.max())

```

Now, remove all dashes from the full sequences so that we only have the pure sequence without any multi sequence alignment characters

```python
tidied_data['Full sequence'] = tidied_data['Full sequence'].str.replace('-', '', regex=False)
tidied_data['CDR1'] = tidied_data['CDR1'].str.replace('-', '', regex=False)
tidied_data['CDR2'] = tidied_data['CDR2'].str.replace('-', '', regex=False)
tidied_data['CDR3'] = tidied_data['CDR3'].str.replace('-', '', regex=False)
tidied_data['Framework 1'] = tidied_data['Framework 1'].str.replace('-', '', regex=False)
tidied_data['Framework 2'] = tidied_data['Framework 2'].str.replace('-', '', regex=False)
tidied_data['Framework 3'] = tidied_data['Framework 3'].str.replace('-', '', regex=False)
```

Drop all rows where CDR3 does not have a value

```python
tidied_data = tidied_data.dropna(subset=['CDR3'])
```

Drop all rows where there is no value from delphia

```python
tidied_data = tidied_data.dropna(subset=['Ag1_Raw signals'])
# Drop rows where 'CDR3' or 'Full sequence' is NaN
tidied_data = tidied_data.dropna(subset=['CDR3', 'Full sequence'])
```

tidy names

```python
tidied_data.loc[tidied_data['Ag1'].str.contains('SVMP'), 'Ag1_modified'] = 'SVMP'
tidied_data.loc[tidied_data['Ag1'].str.contains('SVSP'), 'Ag1_modified'] = 'SVSP'
tidied_data.loc[tidied_data['Ag1'].str.contains('PLA2'), 'Ag1_modified'] = 'PLA2'
tidied_data.loc[tidied_data['Ag1'].str.contains('Dp8'), 'Ag1_modified'] = 'Dp8'
tidied_data.loc[tidied_data['Ag1'].str.contains('pla2'), 'Ag1_modified'] = 'PLA2'
tidied_data.loc[tidied_data['Ag1'].str.contains('Nu6'), 'Ag1_modified'] = 'Nu6' # discard because it has 50/50 fraction of pla2 and CTx
tidied_data.loc[tidied_data['Ag1'].str.contains('Dp4'), 'Ag1_modified'] = 'Dp4'
tidied_data.loc[tidied_data['Ag1'].str.contains('Nm8'), 'Ag1_modified'] = 'Nm8'
tidied_data.loc[tidied_data['Ag1'].str.contains('Dv4'), 'Ag1_modified'] = 'Dv4' # not listed
tidied_data.loc[tidied_data['Ag1'].str.contains('scNTX'), 'Ag1_modified'] = 'scNTX'
```

```python
tidied_data.to_excel("nanobody_delphia_no_ag_filter.xlsx", index=False)
```

In the next step we rename the targets. Everything which has a higher fraction of 70% in the venom composition will be taken and renamed corresponding to the most abundant fraction.
SVMP, SVSP and PLA2 are kept
NU6 is discarded because it has a 50/50 fraction of pla2 and CTx

```python
tidied_data.loc[tidied_data['Ag1'].str.contains('SVMP'), 'Ag1_modified'] = 'SVMP'
tidied_data.loc[tidied_data['Ag1'].str.contains('SVSP'), 'Ag1_modified'] = 'SVSP'
tidied_data.loc[tidied_data['Ag1'].str.contains('PLA2'), 'Ag1_modified'] = 'PLA2'
tidied_data.loc[tidied_data['Ag1'].str.contains('Dp8'), 'Ag1_modified'] = 'Kunitz'
tidied_data.loc[tidied_data['Ag1'].str.contains('pla2'), 'Ag1_modified'] = 'PLA2'
tidied_data.loc[tidied_data['Ag1'].str.contains('Nu6'), 'Ag1_modified'] = 'Nu6' # discard because it has 50/50 fraction of pla2 and CTx
tidied_data.loc[tidied_data['Ag1'].str.contains('Dp4'), 'Ag1_modified'] = 'sNTx'
tidied_data.loc[tidied_data['Ag1'].str.contains('Nm8'), 'Ag1_modified'] = 'INTx'
tidied_data.loc[tidied_data['Ag1'].str.contains('Dv4'), 'Ag1_modified'] = 'Dv4' # not listed
tidied_data.loc[tidied_data['Ag1'].str.contains('scNTX'), 'Ag1_modified'] = 'scNTX'
tidied_data = tidied_data.dropna(subset=['Ag1_modified'])
tidied_data = tidied_data[tidied_data['Ag1_modified'] != 'Dv4']
tidied_data = tidied_data[tidied_data['Ag1_modified'] != 'Nu6']
tidied_data = tidied_data[tidied_data['Ag1_modified'] != 'scNTX']
```

```python
tidied_data.to_excel("nanobody_delphia_tidied.xlsx", index=False)
```

Make fasta out of sequences

```python

fasta_file = "nanobody_sequences.fasta"

codon_table = {
    'A': 'GCT', 'R': 'CGT', 'N': 'AAT', 'D': 'GAT', 'C': 'TGT', 
    'Q': 'CAA', 'E': 'GAA', 'G': 'GGT', 'H': 'CAT', 'I': 'ATT', 
    'L': 'CTT', 'K': 'AAA', 'M': 'ATG', 'F': 'TTT', 'P': 'CCT', 
    'S': 'TCT', 'T': 'ACT', 'W': 'TGG', 'Y': 'TAT', 'V': 'GTT', 
    '*': 'TGA'  
}

with open(fasta_file, 'w') as file:
    for idx, row in tidied_data.iterrows():
        amino_acid_seq = row['Full sequence']
        nucleotide_seq = ''.join(codon_table[aa] for aa in amino_acid_seq if aa in codon_table)
        file.write(f">" + f"sequence{idx}" + "\n")
        file.write(str(nucleotide_seq) + "\n")
```
